```python
import torch
import gc

def clear_gpu_memory():
    """
    PyTorch GPU 메모리 캐시를 완전히 초기화하는 함수.
    - garbage collection 강제 실행
    - CUDA 캐시 해제
    - peak memory stats 리셋 (디버깅용)
    """
    # Python 객체 참조 해제
    gc.collect()
    
    # PyTorch CUDA 캐시 해제
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.synchronize()  # 동기화로 pending operation 완료 보장
        torch.cuda.reset_peak_memory_stats()  # peak 메모리 통계 리셋 (옵션)
        print(f"GPU memory cleared. Current allocated: {torch.cuda.memory_allocated() / 1024**3:.2f} GB")
        print(f"GPU memory reserved: {torch.cuda.memory_reserved() / 1024**3:.2f} GB")
    else:
        print("CUDA not available. Running on CPU.")

# 사용 예시
clear_gpu_memory()
```

```python
# ============================================================
# Surface-consistent FVM + FVM-guided PINN (RAR + Adam->L-BFGS)
# 1D transient heat conduction in cylindrical wall (r only)
# + Q_REF split (inner/outer) + GIF exports (1D + 2D annulus)
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

import torch
import torch.nn as nn
import torch.autograd as autograd
import random
from dataclasses import dataclass
from typing import Callable, Dict, Optional, Tuple

# --- GIF related (added) ---
from matplotlib import animation
from matplotlib.animation import PillowWriter
from matplotlib.patches import Circle

# -----------------------------
# 0) Global numerics & helpers
# -----------------------------
torch.set_default_dtype(torch.float64)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SIGMA = 5.670374419e-8  # Stefan–Boltzmann [W/m^2/K^4]

def fix_seed(seed: int = 2025):
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

def _to_np(x):
    if isinstance(x, torch.Tensor):
        return x.detach().cpu().numpy()
    return x

# -----------------------------
# 1) Material models
# -----------------------------
class MaterialModel:
    """k(T), cp(T) + rho"""
    def __init__(self, rho: float):
        self.rho = float(rho)

    def k_np(self, T: np.ndarray) -> np.ndarray:
        raise NotImplementedError

    def cp_np(self, T: np.ndarray) -> np.ndarray:
        raise NotImplementedError

    def k_t(self, T: torch.Tensor) -> torch.Tensor:
        raise NotImplementedError

    def cp_t(self, T: torch.Tensor) -> torch.Tensor:
        raise NotImplementedError


class ConstMaterial(MaterialModel):
    def __init__(self, rho: float, k: float, cp: float):
        super().__init__(rho=rho)
        self.k0 = float(k)
        self.cp0 = float(cp)

    def k_np(self, T: np.ndarray) -> np.ndarray:
        return np.full_like(T, self.k0, dtype=float)

    def cp_np(self, T: np.ndarray) -> np.ndarray:
        return np.full_like(T, self.cp0, dtype=float)

    def k_t(self, T: torch.Tensor) -> torch.Tensor:
        return torch.full_like(T, self.k0)

    def cp_t(self, T: torch.Tensor) -> torch.Tensor:
        return torch.full_like(T, self.cp0)


class IN718Linear(MaterialModel):
    """Linearized IN718 approximations (example)."""
    def __init__(self, rho: float = 8200.0):
        super().__init__(rho=rho)

    def k_np(self, T: np.ndarray) -> np.ndarray:
        return 15.0 - 0.004 * (T - 300.0)

    def cp_np(self, T: np.ndarray) -> np.ndarray:
        return 435.0 + 0.25 * (T - 300.0)

    def k_t(self, T: torch.Tensor) -> torch.Tensor:
        return 15.0 - 0.004 * (T - 300.0)

    def cp_t(self, T: torch.Tensor) -> torch.Tensor:
        return 435.0 + 0.25 * (T - 300.0)

# -----------------------------
# 2) Boundary condition models
# -----------------------------
@dataclass
class InnerBC:
    """q_in = h(t) * (T_inf(t) - T_wall_inner)"""
    h_fun_np: Callable[[float], float]
    h_fun_t: Callable[[torch.Tensor], torch.Tensor]
    Tinf_fun_np: Callable[[float], float]
    Tinf_fun_t: Callable[[torch.Tensor], torch.Tensor]

@dataclass
class OuterBC:
    """
    q_out = h_out*(T_wall_outer - T_amb) + eps*sigma*(T_wall_outer^4 - T_amb^4)
    For implicit FVM, we use Newton linearization of radiation term about current Tw.
    """
    h_out: float
    eps: float
    T_amb: float

    def q_env_np(self, Tw: float) -> float:
        q_conv = self.h_out * (Tw - self.T_amb)
        q_rad  = self.eps * SIGMA * (Tw**4 - self.T_amb**4)
        return q_conv + q_rad

    def q_env_t(self, Tw: torch.Tensor) -> torch.Tensor:
        q_conv = self.h_out * (Tw - self.T_amb)
        q_rad  = self.eps * SIGMA * (Tw**4 - (self.T_amb**4))
        return q_conv + q_rad

    def linearize_rad_np(self, Tw_star: float) -> Tuple[float, float]:
        """
        Radiation linearization:
          eps*sigma*(Tw^4 - Tamb^4) ≈ h_rad*(Tw - T_lin)
        with:
          h_rad = 4 eps sigma Tw_star^3
          T_lin = Tw_star + (Tamb^4 - Tw_star^4)/(4 Tw_star^3)
        """
        if self.eps <= 0.0:
            return 0.0, self.T_amb
        Tws = max(float(Tw_star), 1.0)  # avoid 0
        h_rad = 4.0 * self.eps * SIGMA * (Tws**3)
        T_lin = Tws + (self.T_amb**4 - Tws**4) / (4.0 * (Tws**3))
        return h_rad, T_lin

# -----------------------------
# 3) Case params
# -----------------------------
@dataclass
class CaseParams:
    name: str
    Ri: float             # inner radius [m]
    Lw: float             # wall thickness [m]
    t_end: float          # [s]
    T_init: float         # [K]
    material: MaterialModel
    inner: InnerBC
    outer: OuterBC

    # FVM numerics
    fvm_Nr: int = 120
    fvm_dt: float = 1e-4
    fvm_inner_iters: int = 30
    fvm_inner_tol: float = 5e-3
    fvm_relax: float = 0.6

    # PINN numerics
    use_teacher: bool = True

@dataclass
class TrainConfig:
    # Adam stage
    adam_epochs: int = 3000
    lr_adam: float = 1e-3
    print_every: int = 500

    # Collocation sizes (Adam)
    N_int: int = 6000
    N_bc: int  = 3000
    N_ic: int  = 1500
    N_data: int = 4000

    # time curriculum
    curriculum_warm: int = 2000
    bias_power: float = 2.0

    # teacher weight schedule
    W_DATA0: float = 20.0
    teacher_off_epoch: int = 2500

    # main loss weights
    W_PDE: float = 1.0
    W_IC: float = 1.0
    W_BC_IN: float = 20.0
    W_BC_OUT: float = 5.0

    # RAR (residual-based adaptive refinement)
    use_rar: bool = True
    rar_period: int = 200
    rar_candidates: int = 20000
    rar_add: int = 2000

    # L-BFGS stage
    use_lbfgs: bool = True
    lbfgs_max_iter: int = 500
    lbfgs_history: int = 50
    lbfgs_lr: float = 1.0

    # fixed dataset sizes for L-BFGS
    N_int_fixed: int = 30000
    N_bc_fixed: int  = 10000
    N_ic_fixed: int  = 5000
    N_data_fixed: int = 20000

# -----------------------------
# 4) FVM solver (surface-consistent)
# -----------------------------
def run_fvm(case: CaseParams) -> Dict[str, np.ndarray]:
    """
    Cell-centered FVM in r with implicit time stepping.
    Stores:
      - T_hist at cell centers
      - Tw_in_hist, Tw_out_hist (surface temperatures)
      - q_in_hist, q_out_hist (boundary fluxes using Tw)
    """
    Ri = case.Ri
    Ro = case.Ri + case.Lw
    N  = int(case.fvm_Nr)
    dt = float(case.fvm_dt)
    t_end = float(case.t_end)

    r_faces   = np.linspace(Ri, Ro, N+1)
    r_centers = 0.5 * (r_faces[:-1] + r_faces[1:])

    # per-unit-length areas/volumes
    A_face = 2.0 * np.pi * r_faces
    V_cell = np.pi * (r_faces[1:]**2 - r_faces[:-1]**2)

    dcc   = r_centers[1:] - r_centers[:-1]
    dr_w0 = r_centers[0] - r_faces[0]
    dr_eN = r_faces[-1] - r_centers[-1]

    T = np.full(N, case.T_init, dtype=float)

    steps = int(np.ceil(t_end / dt))
    t_hist = np.zeros(steps+1)
    T_hist = np.zeros((steps+1, N))
    Tw_in_hist  = np.zeros(steps+1)
    Tw_out_hist = np.zeros(steps+1)
    q_in_hist   = np.zeros(steps+1)
    q_out_hist  = np.zeros(steps+1)

    Tw_in_hist[0]  = case.T_init
    Tw_out_hist[0] = case.T_init
    q_in_hist[0]   = 0.0
    q_out_hist[0]  = 0.0
    T_hist[0] = T.copy()

    for n in range(1, steps+1):
        t_now = n * dt
        t_hist[n] = t_now
        T_prev = T.copy()
        T_iter = T.copy()

        TwL_it = Tw_in_hist[n-1]
        TwR_it = Tw_out_hist[n-1]

        for it in range(case.fvm_inner_iters):
            k_cell  = case.material.k_np(T_iter)
            cp_cell = case.material.cp_np(T_iter)
            Ccap = case.material.rho * cp_cell * V_cell  # [J/K] per unit length

            # harmonic mean k at internal faces
            k_face = np.empty(N+1)
            k_face[0]  = k_cell[0]
            k_face[-1] = k_cell[-1]
            for j in range(1, N):
                kL, kR = k_cell[j-1], k_cell[j]
                k_face[j] = 2.0 * kL * kR / max(kL + kR, 1e-30)

            G_int = k_face[1:N] * A_face[1:N] / dcc
            aW = np.zeros(N); aE = np.zeros(N)
            aW[1:]  = G_int
            aE[:-1] = G_int

            aP = aW + aE + Ccap/dt
            b  = (Ccap/dt) * T_prev

            # --- Inner Robin ---
            h_in = case.inner.h_fun_np(t_now)
            Tinf_in = case.inner.Tinf_fun_np(t_now)

            kL_eff = k_face[0] / max(dr_w0, 1e-30)
            H_L = (kL_eff * h_in) / max(kL_eff + h_in, 1e-30)

            # --- Outer conv + rad (linearized) ---
            h_conv = case.outer.h_out
            h_rad, T_lin = case.outer.linearize_rad_np(TwR_it)
            h_sum = h_conv + h_rad
            if h_sum > 0.0:
                T_eff = (h_conv*case.outer.T_amb + h_rad*T_lin) / h_sum
            else:
                T_eff = case.outer.T_amb

            kR_eff = k_face[-1] / max(dr_eN, 1e-30)
            H_R = (kR_eff * h_sum) / max(kR_eff + h_sum, 1e-30)

            # apply to matrix
            aP[0]  += H_L * A_face[0]
            b[0]   += H_L * A_face[0] * Tinf_in

            aP[-1] += H_R * A_face[-1]
            b[-1]  += H_R * A_face[-1] * T_eff

            # --- TDMA ---
            lower = -aW
            diag  = aP
            upper = -aE

            c = np.zeros(N)
            d = np.zeros(N)

            c[0] = upper[0] / diag[0]
            d[0] = b[0] / diag[0]
            for i in range(1, N):
                denom = diag[i] - lower[i] * c[i-1]
                if i < N-1:
                    c[i] = upper[i] / denom
                d[i] = (b[i] - lower[i] * d[i-1]) / denom

            T_new = np.empty_like(T_iter)
            T_new[-1] = d[-1]
            for i in range(N-2, -1, -1):
                T_new[i] = d[i] - c[i] * T_new[i+1]

            # surface temps
            TwL_new = (kL_eff*T_new[0] + h_in*Tinf_in) / max(kL_eff + h_in, 1e-30)
            TwR_new = (kR_eff*T_new[-1] + h_sum*T_eff) / max(kR_eff + h_sum, 1e-30)

            upd_T  = np.max(np.abs(T_new - T_iter))
            upd_Tw = max(abs(TwL_new - TwL_it), abs(TwR_new - TwR_it))

            alpha = float(case.fvm_relax)
            T_iter = T_iter + alpha * (T_new - T_iter)
            TwL_it = TwL_it + alpha * (TwL_new - TwL_it)
            TwR_it = TwR_it + alpha * (TwR_new - TwR_it)

            if upd_T < case.fvm_inner_tol and upd_Tw < case.fvm_inner_tol:
                break

        T = T_iter.copy()
        T_hist[n] = T.copy()
        Tw_in_hist[n]  = TwL_it
        Tw_out_hist[n] = TwR_it

        q_in  = case.inner.h_fun_np(t_now) * (case.inner.Tinf_fun_np(t_now) - Tw_in_hist[n])
        q_out = case.outer.q_env_np(Tw_out_hist[n])
        q_in_hist[n]  = q_in
        q_out_hist[n] = q_out

    return {
        "t": t_hist,
        "r_centers": r_centers,
        "T_centers": T_hist,
        "Tw_in": Tw_in_hist,
        "Tw_out": Tw_out_hist,
        "q_in": q_in_hist,
        "q_out": q_out_hist,
        "Ri": np.array([Ri]),
        "Ro": np.array([Ro]),
    }

# -----------------------------
# 5) Teacher interpolation
# -----------------------------
class FVMTeacher:
    def __init__(self, fvm: Dict[str, np.ndarray], device=device):
        self.t = torch.tensor(fvm["t"], device=device)
        self.r = torch.tensor(fvm["r_centers"], device=device)
        self.T = torch.tensor(fvm["T_centers"], device=device)  # [Nt, Nr]
        self.Tw_in  = torch.tensor(fvm["Tw_in"], device=device)  # [Nt]
        self.Tw_out = torch.tensor(fvm["Tw_out"], device=device) # [Nt]
        self.q_in   = torch.tensor(fvm["q_in"], device=device)
        self.q_out  = torch.tensor(fvm["q_out"], device=device)
        self.Ri = float(fvm["Ri"][0])
        self.Ro = float(fvm["Ro"][0])

    @torch.no_grad()
    def interp_T(self, r_s: torch.Tensor, t_s: torch.Tensor) -> torch.Tensor:
        r = torch.clamp(r_s.view(-1), self.r[0], self.r[-1])
        t = torch.clamp(t_s.view(-1), self.t[0], self.t[-1])

        Nr = self.r.shape[0]
        Nt = self.t.shape[0]

        ir = torch.searchsorted(self.r, r) - 1
        ir = torch.clamp(ir, 0, Nr-2)
        ir2 = ir + 1

        it = torch.searchsorted(self.t, t) - 1
        it = torch.clamp(it, 0, Nt-2)
        it2 = it + 1

        r1 = self.r[ir]; r2 = self.r[ir2]
        t1 = self.t[it]; t2 = self.t[it2]

        wr = (r - r1) / torch.clamp(r2 - r1, 1e-12)
        wt = (t - t1) / torch.clamp(t2 - t1, 1e-12)

        T11 = self.T[it,  ir]
        T12 = self.T[it2, ir]
        T21 = self.T[it,  ir2]
        T22 = self.T[it2, ir2]

        T_r1 = T11 * (1-wt) + T12 * wt
        T_r2 = T21 * (1-wt) + T22 * wt
        Tout = T_r1 * (1-wr) + T_r2 * wr
        return Tout.view(-1, 1)

    @torch.no_grad()
    def interp_Tw(self, which: str, t_s: torch.Tensor) -> torch.Tensor:
        t = torch.clamp(t_s.view(-1), self.t[0], self.t[-1])
        Nt = self.t.shape[0]
        it = torch.searchsorted(self.t, t) - 1
        it = torch.clamp(it, 0, Nt-2)
        it2 = it + 1
        t1 = self.t[it]; t2 = self.t[it2]
        w = (t - t1) / torch.clamp(t2 - t1, 1e-12)

        if which == "in":
            y1 = self.Tw_in[it]; y2 = self.Tw_in[it2]
        else:
            y1 = self.Tw_out[it]; y2 = self.Tw_out[it2]
        y = y1*(1-w) + y2*w
        return y.view(-1, 1)

# -----------------------------
# 6) PINN model (IC hard constraint)
# -----------------------------
class PINNNet(nn.Module):
    def __init__(self, hidden_dim=64, n_layers=6):
        super().__init__()
        layers = []
        in_dim = 2
        for _ in range(n_layers):
            layers += [nn.Linear(in_dim, hidden_dim), nn.Tanh()]
            in_dim = hidden_dim
        layers += [nn.Linear(hidden_dim, 1)]
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)

class PINNModel(nn.Module):
    """
    Hard IC:
      T(r,0) = T_init
    using:
      s(t) = 1 - exp(-beta * t_nd)
      T = T_init + s(t) * T_char * sigmoid(net([r_nd,t_nd]))
    """
    def __init__(self, case: CaseParams, hidden_dim=64, n_layers=6, beta_ic=5.0):
        super().__init__()
        self.case = case
        self.Ri = float(case.Ri)
        self.Ro = float(case.Ri + case.Lw)
        self.t_end = float(case.t_end)
        self.T_init = float(case.T_init)

        ts = np.linspace(0.0, self.t_end, 50)
        Tinf_vals = np.array([case.inner.Tinf_fun_np(float(tt)) for tt in ts])
        T_char = max(float(np.max(Tinf_vals) - self.T_init), 1.0)
        self.T_char = float(T_char)

        self.beta_ic = float(beta_ic)
        self.net = PINNNet(hidden_dim=hidden_dim, n_layers=n_layers)

    def nd_r(self, r):
        return (r - self.Ri) / (self.Ro - self.Ri)

    def nd_t(self, t):
        return t / self.t_end

    def forward(self, r_phys: torch.Tensor, t_phys: torch.Tensor) -> torch.Tensor:
        r_nd = self.nd_r(r_phys)
        t_nd = self.nd_t(t_phys)
        x = torch.cat([r_nd, t_nd], dim=1)
        x = 2.0*x - 1.0
        y = self.net(x)
        theta = torch.sigmoid(y)
        s = 1.0 - torch.exp(-self.beta_ic * t_nd)
        T = self.T_init + s * (self.T_char * theta)
        return T

# -----------------------------
# 7) Sampling utilities (+ curriculum)
# -----------------------------
def sample_interior(case: CaseParams, N: int, time_frac: float, bias_power: float):
    Ri = case.Ri
    Ro = case.Ri + case.Lw
    r = torch.rand(N, 1, device=device) * (Ro - Ri) + Ri
    u = torch.rand(N, 1, device=device)
    t = (u**bias_power) * (time_frac * case.t_end)
    r.requires_grad_(True)
    t.requires_grad_(True)
    return r, t

def sample_boundary(case: CaseParams, N: int, time_frac: float, bias_power: float):
    u = torch.rand(N, 1, device=device)
    t = (u**bias_power) * (time_frac * case.t_end)
    r_in  = torch.full_like(t, case.Ri)
    r_out = torch.full_like(t, case.Ri + case.Lw)
    r_in.requires_grad_(True)
    r_out.requires_grad_(True)
    t.requires_grad_(True)
    return r_in, r_out, t

def sample_initial(case: CaseParams, N: int):
    Ri = case.Ri
    Ro = case.Ri + case.Lw
    r = torch.rand(N, 1, device=device) * (Ro - Ri) + Ri
    t = torch.zeros(N, 1, device=device)
    r.requires_grad_(True)
    t.requires_grad_(True)
    return r, t

# -----------------------------
# 8) Physics residuals and loss terms
# -----------------------------
def build_scales(case: CaseParams, model: PINNModel) -> Dict[str, torch.Tensor]:
    """
    (MOD) Q_REF split:
      - Q_REF_IN : scale for inner Robin BC flux
      - Q_REF_OUT: scale for outer env flux (conv+rad), to prevent being dwarfed by q_in
    """
    # residual scale (K/s)
    R_ref = max(model.T_char / max(case.t_end, 1e-12), 1e-6)

    # inner flux scale ~ max(h_in)*T_char
    ts = np.linspace(0.0, case.t_end, 60)
    hmax_in = max([case.inner.h_fun_np(float(tt)) for tt in ts] + [1.0])
    Q_ref_in = max(hmax_in * model.T_char, 1.0e5)

    # outer flux scale: conservative estimate at Tw_peak = T_init + T_char
    Tw_peak = float(case.T_init + model.T_char)
    q_out_peak = abs(case.outer.h_out * (Tw_peak - case.outer.T_amb) +
                     case.outer.eps * SIGMA * (Tw_peak**4 - case.outer.T_amb**4))
    # floor to avoid too-small scale (unstable overweight)
    Q_ref_out = max(q_out_peak, 1.0e4)

    return {
        "R_REF": torch.tensor(R_ref, device=device),
        "Q_REF_IN": torch.tensor(Q_ref_in, device=device),
        "Q_REF_OUT": torch.tensor(Q_ref_out, device=device),
        "T_CHAR": torch.tensor(model.T_char, device=device),
    }

def pde_residual(case: CaseParams, model: PINNModel, scales: Dict[str, torch.Tensor],
                 r: torch.Tensor, t: torch.Tensor) -> torch.Tensor:
    """
    PDE:
      dT/dt = 1/(rho*cp(T)) * (1/r) d/dr ( k(T) r dT/dr )
    """
    T = model(r, t)

    T_t = autograd.grad(T, t, grad_outputs=torch.ones_like(T),
                        retain_graph=True, create_graph=True)[0]
    T_r = autograd.grad(T, r, grad_outputs=torch.ones_like(T),
                        retain_graph=True, create_graph=True)[0]

    k  = case.material.k_t(T)
    cp = case.material.cp_t(T)
    rho = torch.tensor(case.material.rho, device=device)

    qr = k * r * T_r
    qr_r = autograd.grad(qr, r, grad_outputs=torch.ones_like(qr),
                         retain_graph=True, create_graph=True)[0]
    rhs = (1.0 / (rho * cp)) * (1.0 / r) * qr_r

    res = (T_t - rhs) / scales["R_REF"]
    return res

def ic_residual(case: CaseParams, model: PINNModel, scales: Dict[str, torch.Tensor],
                r0: torch.Tensor, t0: torch.Tensor) -> torch.Tensor:
    T = model(r0, t0)
    return (T - case.T_init) / scales["T_CHAR"]

def bc_inner_residual(case: CaseParams, model: PINNModel, scales: Dict[str, torch.Tensor],
                      r_in: torch.Tensor, t: torch.Tensor) -> torch.Tensor:
    Tw = model(r_in, t)
    Tw_r = autograd.grad(Tw, r_in, grad_outputs=torch.ones_like(Tw),
                         retain_graph=True, create_graph=True)[0]
    k = case.material.k_t(Tw)
    q_cond = -k * Tw_r  # + into wall if Tw decreases outward

    h = case.inner.h_fun_t(t)
    Tinf = case.inner.Tinf_fun_t(t)
    q_bc = h * (Tinf - Tw)

    # (MOD) inner scale
    return (q_cond - q_bc) / scales["Q_REF_IN"]

def bc_outer_residual(case: CaseParams, model: PINNModel, scales: Dict[str, torch.Tensor],
                      r_out: torch.Tensor, t: torch.Tensor) -> torch.Tensor:
    Tw = model(r_out, t)
    Tw_r = autograd.grad(Tw, r_out, grad_outputs=torch.ones_like(Tw),
                         retain_graph=True, create_graph=True)[0]
    k = case.material.k_t(Tw)
    q_cond = -k * Tw_r

    q_env = case.outer.q_env_t(Tw)

    # (MOD) outer scale
    return (q_cond - q_env) / scales["Q_REF_OUT"]

def teacher_residual(case: CaseParams, model: PINNModel, scales: Dict[str, torch.Tensor],
                     teacher: FVMTeacher, N: int, time_frac: float) -> torch.Tensor:
    r, t = sample_interior(case, N, time_frac=time_frac, bias_power=1.0)
    with torch.no_grad():
        T_f = teacher.interp_T(r, t)
    T_p = model(r, t)
    return (T_p - T_f) / scales["T_CHAR"]

def teacher_surface_residual(case: CaseParams, model: PINNModel, scales: Dict[str, torch.Tensor],
                             teacher: FVMTeacher, N: int, time_frac: float) -> Tuple[torch.Tensor, torch.Tensor]:
    u = torch.rand(N, 1, device=device)
    t = (u**1.0) * (time_frac * case.t_end)
    t.requires_grad_(False)

    r_in = torch.full_like(t, case.Ri)
    r_out = torch.full_like(t, case.Ri + case.Lw)

    with torch.no_grad():
        Tw_in_f  = teacher.interp_Tw("in", t)
        Tw_out_f = teacher.interp_Tw("out", t)

    Tw_in_p  = model(r_in, t)
    Tw_out_p = model(r_out, t)

    res_in  = (Tw_in_p  - Tw_in_f)  / scales["T_CHAR"]
    res_out = (Tw_out_p - Tw_out_f) / scales["T_CHAR"]
    return res_in, res_out

# -----------------------------
# 9) RAR: residual-based adaptive refinement
# -----------------------------
def rar_select_points(case: CaseParams, model: PINNModel, scales: Dict[str, torch.Tensor],
                      candidates: int, add: int) -> Tuple[torch.Tensor, torch.Tensor]:
    """
    NOTE: keep this as-is structurally.
    (You can later optimize memory by disabling param grads during RAR selection,
     but for now we keep minimal changes.)
    """
    r_c, t_c = sample_interior(case, candidates, time_frac=1.0, bias_power=1.0)
    r_c.requires_grad_(True)
    t_c.requires_grad_(True)
    res = pde_residual(case, model, scales, r_c, t_c).detach().abs().view(-1)
    idx = torch.topk(res, k=min(add, res.numel()), largest=True).indices
    r_sel = r_c.detach()[idx].clone().requires_grad_(True)
    t_sel = t_c.detach()[idx].clone().requires_grad_(True)
    return r_sel, t_sel

# -----------------------------
# 10) Training
# -----------------------------
def train(case: CaseParams,
          cfg: TrainConfig,
          teacher: Optional[FVMTeacher]) -> Tuple[PINNModel, Dict[str, list]]:
    fix_seed(2025)

    model = PINNModel(case, hidden_dim=64, n_layers=6, beta_ic=5.0).to(device)
    scales = build_scales(case, model)
    opt = torch.optim.Adam(model.parameters(), lr=cfg.lr_adam)

    history = {"loss": [], "pde": [], "ic": [], "bci": [], "bco": [], "data": [], "wd": []}

    rar_r = None
    rar_t = None

    for ep in range(1, cfg.adam_epochs + 1):
        opt.zero_grad()

        if cfg.curriculum_warm > 0:
            time_frac = float(min(1.0, 0.1 + 0.9 * ep / cfg.curriculum_warm))
        else:
            time_frac = 1.0
        bias_p = float(cfg.bias_power)

        if cfg.use_rar and (ep % cfg.rar_period == 0):
            try:
                rar_r, rar_t = rar_select_points(
                    case, model, scales,
                    candidates=cfg.rar_candidates, add=cfg.rar_add
                )
            except RuntimeError as e:
                print(f"[RAR skipped] {e}")
                rar_r, rar_t = None, None

        # PDE
        r_int, t_int = sample_interior(case, cfg.N_int, time_frac, bias_p)
        res_pde = pde_residual(case, model, scales, r_int, t_int)
        loss_pde = torch.mean(res_pde**2)

        if rar_r is not None and rar_t is not None:
            res_rar = pde_residual(case, model, scales, rar_r, rar_t)
            loss_pde = 0.5 * loss_pde + 0.5 * torch.mean(res_rar**2)

        # IC
        r0, t0 = sample_initial(case, cfg.N_ic)
        res_ic = ic_residual(case, model, scales, r0, t0)
        loss_ic = torch.mean(res_ic**2)

        # BCs
        r_in, r_out, t_bc = sample_boundary(case, cfg.N_bc, time_frac, bias_p)
        res_bci = bc_inner_residual(case, model, scales, r_in, t_bc)
        res_bco = bc_outer_residual(case, model, scales, r_out, t_bc)
        loss_bci = torch.mean(res_bci**2)
        loss_bco = torch.mean(res_bco**2)

        # teacher
        if (case.use_teacher and (teacher is not None)):
            res_data = teacher_residual(case, model, scales, teacher, cfg.N_data, time_frac)
            loss_data = torch.mean(res_data**2)

            res_s_in, res_s_out = teacher_surface_residual(case, model, scales, teacher,
                                                          N=max(500, cfg.N_data//8),
                                                          time_frac=time_frac)
            loss_data = 0.7*loss_data + 0.15*torch.mean(res_s_in**2) + 0.15*torch.mean(res_s_out**2)

            if cfg.teacher_off_epoch > 0:
                decay = max(0.0, 1.0 - ep / cfg.teacher_off_epoch)
            else:
                decay = 0.0
            Wd = cfg.W_DATA0 * decay
        else:
            loss_data = torch.tensor(0.0, device=device)
            Wd = 0.0

        loss = (cfg.W_PDE * loss_pde +
                cfg.W_IC * loss_ic +
                cfg.W_BC_IN * loss_bci +
                cfg.W_BC_OUT * loss_bco +
                Wd * loss_data)

        loss.backward()
        opt.step()

        history["loss"].append(float(loss.detach().cpu()))
        history["pde"].append(float(loss_pde.detach().cpu()))
        history["ic"].append(float(loss_ic.detach().cpu()))
        history["bci"].append(float(loss_bci.detach().cpu()))
        history["bco"].append(float(loss_bco.detach().cpu()))
        history["data"].append(float(loss_data.detach().cpu()))
        history["wd"].append(float(Wd))

        if (ep == 1) or (ep % cfg.print_every == 0):
            print(f"[{case.name} | Adam {ep:5d}] "
                  f"L={loss.item():.3e} PDE={loss_pde.item():.3e} IC={loss_ic.item():.3e} "
                  f"BCi={loss_bci.item():.3e} BCo={loss_bco.item():.3e} "
                  f"DATA={loss_data.item():.3e} Wd={Wd:.2f} t_frac={time_frac:.2f}")

    # ---- L-BFGS refinement on fixed dataset ----
    if cfg.use_lbfgs:
        r_int_f, t_int_f = sample_interior(case, cfg.N_int_fixed, time_frac=1.0, bias_power=1.0)
        r_in_f, r_out_f, t_bc_f = sample_boundary(case, cfg.N_bc_fixed, time_frac=1.0, bias_power=1.0)
        r0_f, t0_f = sample_initial(case, cfg.N_ic_fixed)

        if (case.use_teacher and (teacher is not None)):
            r_data_f, t_data_f = sample_interior(case, cfg.N_data_fixed, time_frac=1.0, bias_power=1.0)
            with torch.no_grad():
                T_data_f = teacher.interp_T(r_data_f, t_data_f)
        else:
            r_data_f, t_data_f, T_data_f = None, None, None

        lbfgs = torch.optim.LBFGS(
            model.parameters(),
            lr=cfg.lbfgs_lr,
            max_iter=cfg.lbfgs_max_iter,
            history_size=cfg.lbfgs_history,
            line_search_fn="strong_wolfe",
        )

        def closure():
            lbfgs.zero_grad()

            lpde = torch.mean(pde_residual(case, model, scales, r_int_f, t_int_f)**2)
            lic  = torch.mean(ic_residual(case, model, scales, r0_f, t0_f)**2)
            lbci = torch.mean(bc_inner_residual(case, model, scales, r_in_f, t_bc_f)**2)
            lbco = torch.mean(bc_outer_residual(case, model, scales, r_out_f, t_bc_f)**2)

            if (case.use_teacher and (teacher is not None)):
                Tp = model(r_data_f, t_data_f)
                ldata = torch.mean(((Tp - T_data_f) / scales["T_CHAR"])**2)
                Wd = 0.0
            else:
                ldata = torch.tensor(0.0, device=device)
                Wd = 0.0

            L = (cfg.W_PDE*lpde +
                 cfg.W_IC*lic +
                 cfg.W_BC_IN*lbci +
                 cfg.W_BC_OUT*lbco +
                 Wd*ldata)
            L.backward()
            return L

        print(f"[{case.name}] L-BFGS start (max_iter={cfg.lbfgs_max_iter}) ...")
        lbfgs.step(closure)
        print(f"[{case.name}] L-BFGS done.")

    return model, history

# -----------------------------
# 11) Evaluation utilities
# -----------------------------
@torch.no_grad()
def eval_wall_time_series(case: CaseParams, model: PINNModel, Nt: int = 600) -> Dict[str, np.ndarray]:
    t = torch.linspace(0.0, case.t_end, Nt, device=device).view(-1, 1)
    r_in  = torch.full_like(t, case.Ri)
    r_out = torch.full_like(t, case.Ri + case.Lw)

    Tw_in  = model(r_in, t)
    Tw_out = model(r_out, t)

    h_in = case.inner.h_fun_t(t)
    Tinf = case.inner.Tinf_fun_t(t)
    q_in = h_in * (Tinf - Tw_in)

    q_out = case.outer.q_env_t(Tw_out)

    return {
        "t": _to_np(t).reshape(-1),
        "Tw_in": _to_np(Tw_in).reshape(-1),
        "Tw_out": _to_np(Tw_out).reshape(-1),
        "q_in": _to_np(q_in).reshape(-1),
        "q_out": _to_np(q_out).reshape(-1),
    }

def plot_wall_compare(case: CaseParams,
                      fvm: Optional[Dict[str, np.ndarray]],
                      pinn_ts: Dict[str, np.ndarray],
                      title_suffix: str = ""):
    plt.figure(figsize=(7.4, 4.0))
    if fvm is not None:
        plt.plot(fvm["t"], fvm["Tw_in"],  "--", label="FVM inner (surface)")
        plt.plot(fvm["t"], fvm["Tw_out"], "--", label="FVM outer (surface)")
    plt.plot(pinn_ts["t"], pinn_ts["Tw_in"],  label="PINN inner (r=Ri)")
    plt.plot(pinn_ts["t"], pinn_ts["Tw_out"], label="PINN outer (r=Ro)")
    plt.xlabel("time [s]")
    plt.ylabel("Temperature [K]")
    plt.title(f"[Surface-consistent] Wall temperatures: {case.name}{title_suffix}")
    ax = plt.gca()
    ax.xaxis.set_major_locator(MultipleLocator(max(case.t_end/6, 1.0)))
    ax.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_flux(case: CaseParams, fvm: Optional[Dict[str, np.ndarray]], pinn_ts: Dict[str, np.ndarray]):
    plt.figure(figsize=(7.4, 4.0))
    if fvm is not None:
        plt.plot(fvm["t"], fvm["q_in"],  "--", label="FVM q_in")
        plt.plot(fvm["t"], fvm["q_out"], "--", label="FVM q_out")
    plt.plot(pinn_ts["t"], pinn_ts["q_in"],  label="PINN q_in")
    plt.plot(pinn_ts["t"], pinn_ts["q_out"], label="PINN q_out")
    plt.xlabel("time [s]")
    plt.ylabel("Heat flux [W/m^2]")
    plt.title(f"Wall heat flux: {case.name}")
    ax = plt.gca()
    ax.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

def energy_balance_check(case: CaseParams,
                         fvm: Optional[Dict[str, np.ndarray]],
                         pinn_ts: Optional[Dict[str, np.ndarray]] = None,
                         model: Optional[PINNModel] = None,
                         verbose: bool = True):
    """
    Per-unit-length energy balance check:
      dU/dt ≈ A_in*q_in - A_out*q_out

    NOTE:
      - FVM: uses cell-centered T, cp(T)
      - PINN: if pinn_ts provided, must also pass model to sample T(r,t)
    """
    Ri = case.Ri
    Ro = case.Ri + case.Lw
    A_in  = 2.0*np.pi*Ri
    A_out = 2.0*np.pi*Ro

    if fvm is not None:
        t = fvm["t"]
        T = fvm["T_centers"]
        r_faces = np.linspace(Ri, Ro, T.shape[1]+1)
        V_cell = np.pi*(r_faces[1:]**2 - r_faces[:-1]**2)

        U = []
        for k in range(T.shape[0]):
            Tk = T[k]
            cp = case.material.cp_np(Tk)
            U.append(np.sum(case.material.rho * cp * Tk * V_cell))
        U = np.array(U)

        dUdt = np.zeros_like(U)
        dUdt[1:-1] = (U[2:] - U[:-2]) / (t[2:] - t[:-2])
        dUdt[0]    = (U[1] - U[0]) / (t[1] - t[0] + 1e-12)
        dUdt[-1]   = (U[-1] - U[-2]) / (t[-1] - t[-2] + 1e-12)

        rhs = A_in*fvm["q_in"] - A_out*fvm["q_out"]
        rel = np.linalg.norm(dUdt - rhs) / max(np.linalg.norm(rhs), 1e-12)

        if verbose:
            print(f"[Energy balance | FVM] rel_L2(dUdt - (A_in*q_in - A_out*q_out)) = {rel:.3e}")

    if pinn_ts is not None:
        if model is None:
            raise ValueError("energy_balance_check: pinn_ts provided but model is None. Pass model=model.")
        t = pinn_ts["t"]
        Nt = len(t)
        Nr = 200
        r = np.linspace(Ri, Ro, Nr)
        w = np.ones_like(r)
        w[0] *= 0.5; w[-1] *= 0.5
        dr = (Ro - Ri) / (Nr - 1)
        dV = 2.0*np.pi*r*dr*w

        stride = max(1, Nt // 200)
        idxs = np.arange(0, Nt, stride)

        U = np.zeros_like(idxs, dtype=float)
        for j, k in enumerate(idxs):
            tt = torch.tensor(np.full((Nr,1), t[k]), device=device, dtype=torch.get_default_dtype())
            rr = torch.tensor(r.reshape(-1,1), device=device, dtype=torch.get_default_dtype())
            with torch.no_grad():
                Tk = _to_np(model(rr, tt)).reshape(-1)
            cp = case.material.cp_np(Tk)
            U[j] = np.sum(case.material.rho * cp * Tk * dV)

        ts = t[idxs]
        dUdt = np.zeros_like(U)
        dUdt[1:-1] = (U[2:] - U[:-2]) / (ts[2:] - ts[:-2])
        dUdt[0] = (U[1]-U[0])/(ts[1]-ts[0] + 1e-12)
        dUdt[-1] = (U[-1]-U[-2])/(ts[-1]-ts[-2] + 1e-12)

        rhs = A_in*pinn_ts["q_in"][idxs] - A_out*pinn_ts["q_out"][idxs]
        rel = np.linalg.norm(dUdt - rhs) / max(np.linalg.norm(rhs), 1e-12)
        if verbose:
            print(f"[Energy balance | PINN] (coarse) rel_L2 = {rel:.3e}")

# -----------------------------
# 12) Convenience: build common h(t) ramp
# -----------------------------
def make_tanh_ramp_h(h_ref: float, scale: float, tau: float):
    h0 = float(h_ref) * float(scale)
    tau = float(tau)

    def h_np(t: float) -> float:
        x = max(t/tau, 0.0)
        return h0 * np.tanh(x)

    def h_t(t: torch.Tensor) -> torch.Tensor:
        x = torch.clamp(t/tau, min=0.0)
        return torch.tensor(h0, device=t.device) * torch.tanh(x)

    return h_np, h_t

def make_const_T(Tc: float):
    Tc = float(Tc)
    def T_np(t: float) -> float:
        return Tc
    def T_t(t: torch.Tensor) -> torch.Tensor:
        return torch.full_like(t, Tc)
    return T_np, T_t

# -----------------------------
# 13) Presets (A/B)
# -----------------------------
def build_presets() -> Dict[str, CaseParams]:
    Ri = 0.020

    T_hot = 3199.0
    T_hot_np, T_hot_t = make_const_T(T_hot)

    h_ref = 778.72
    h_scale = 1.25
    tau = 0.02
    h_np, h_t = make_tanh_ramp_h(h_ref=h_ref, scale=h_scale, tau=tau)

    inner = InnerBC(
        h_fun_np=h_np,
        h_fun_t=h_t,
        Tinf_fun_np=T_hot_np,
        Tinf_fun_t=T_hot_t,
    )

    T_amb = 280.0
    outer_common = OuterBC(h_out=50.0, eps=0.8, T_amb=T_amb)

    mat_var = IN718Linear(rho=8200.0)
    mat_const = ConstMaterial(rho=8200.0, k=13.0, cp=450.0)

    caseA = CaseParams(
        name="FDM_A_5mm_6s",
        Ri=Ri,
        Lw=0.005,
        t_end=6.0,
        T_init=280.0,
        material=mat_var,      # or mat_const
        inner=inner,
        outer=outer_common,
        fvm_Nr=140,
        fvm_dt=1e-5,
        use_teacher=True
    )

    caseB = CaseParams(
        name="FDM_B_3mm_10s",
        Ri=Ri,
        Lw=0.003,
        t_end=6.0,
        T_init=280.0,
        material=mat_var,      # or mat_const
        inner=inner,
        outer=outer_common,
        fvm_Nr=140,
        fvm_dt=1e-5,
        use_teacher=True
    )

    return {"A": caseA, "B": caseB}

# =====================================================================
# 14-A) GIF 생성: 1D (벽 두께 방향) + 2D 원통 단면  (added)
# =====================================================================
def make_gif_1d_from_pinn(case: CaseParams, model: PINNModel,
                          Nx: int = 120, Nframes: int = 200,
                          path: str = "pinn_1d_temp.gif"):
    """두께 방향 온도 프로파일의 시간 변화를 선그래프로 GIF 출력"""
    Ri = case.Ri
    Ro = case.Ri + case.Lw
    T_END = case.t_end

    t_frames = np.linspace(0.0, T_END, Nframes)
    r_line = torch.linspace(Ri, Ro, Nx, device=device).view(-1, 1)
    x_axis = (r_line.detach().cpu().numpy().flatten() - Ri)

    T_snap = np.zeros((Nframes, Nx), dtype=float)
    for i, tau in enumerate(t_frames):
        t_line = torch.full_like(r_line, float(tau))
        with torch.no_grad():
            T_line = model(r_line, t_line).detach().cpu().numpy().flatten()
        T_snap[i, :] = T_line

    vmin = float(T_snap.min())
    vmax = float(T_snap.max())

    fig, ax = plt.subplots(figsize=(7.0, 3.0))
    line, = ax.plot(x_axis, T_snap[0])
    ax.set_ylim(vmin, vmax)
    ax.set_xlabel("Thickness from inner wall x [m]")
    ax.set_ylabel("Temperature [K]")
    ttl = ax.set_title(f"{case.name} | t={t_frames[0]:.4f} s")
    ax.grid(alpha=0.3)

    def update(i):
        line.set_ydata(T_snap[i])
        ttl.set_text(f"{case.name} | t={t_frames[i]:.4f} s")
        return line, ttl

    fps = 15
    ani = animation.FuncAnimation(fig, update, frames=Nframes,
                                  interval=1000/fps, blit=False)
    ani.save(path, writer=PillowWriter(fps=fps))
    plt.close(fig)
    print(f"Saved 1D GIF: {path}")

def make_gif_2d_cyl_from_pinn(case: CaseParams, model: PINNModel,
                              Nr: int = 120, Ntheta: int = 180,
                              Nframes: int = 120,
                              path: str = "pinn_cyl_temp.gif"):
    """
    원통 단면(annulus)에 온도 분포를 매핑하여 시간 변화 GIF 생성.
    온도는 r만의 함수이므로, θ 방향으로 동일하게 복제.
    """
    Ri = case.Ri
    Ro = case.Ri + case.Lw
    T_END = case.t_end

    t_frames = np.linspace(0.0, T_END, Nframes)
    r_line = torch.linspace(Ri, Ro, Nr, device=device).view(-1, 1)
    r_line_np = r_line.detach().cpu().numpy().flatten()

    T_snap = np.zeros((Nframes, Nr), dtype=float)
    for i, tau in enumerate(t_frames):
        t_line = torch.full_like(r_line, float(tau))
        with torch.no_grad():
            T_line = model(r_line, t_line).detach().cpu().numpy().flatten()
        T_snap[i, :] = T_line

    r_vals = np.linspace(0.0, Ro, Nr)
    theta_vals = np.linspace(0.0, 2.0*np.pi, Ntheta, endpoint=False)
    R, TH = np.meshgrid(r_vals, theta_vals, indexing="ij")
    X = R*np.cos(TH)
    Y = R*np.sin(TH)

    mask_ann = (R >= Ri) & (R <= Ro)

    T2d0 = np.full_like(R, np.nan, dtype=float)
    T2d0[mask_ann] = np.interp(R[mask_ann], r_line_np, T_snap[0])

    vmin = float(np.nanmin(T_snap))
    vmax = float(np.nanmax(T_snap))

    fig, ax = plt.subplots(figsize=(5.0, 5.0))
    pc = ax.pcolormesh(X, Y, T2d0, shading="auto",
                       cmap="viridis", vmin=vmin, vmax=vmax)
    ax.set_aspect("equal")
    ax.set_xlabel("x [m]")
    ax.set_ylabel("y [m]")
    ttl = ax.set_title(f"{case.name} | t={t_frames[0]:.4f} s")
    cbar = fig.colorbar(pc, ax=ax); cbar.set_label("Temperature [K]")

    inner_circle = Circle((0, 0), Ri, facecolor="white",
                          edgecolor="k", lw=0.5, alpha=1.0)
    ax.add_patch(inner_circle)

    def update(i):
        T2d = np.full_like(R, np.nan, dtype=float)
        T2d[mask_ann] = np.interp(R[mask_ann], r_line_np, T_snap[i])
        pc.set_array(T2d.ravel())
        ttl.set_text(f"{case.name} | t={t_frames[i]:.4f} s")
        return pc, ttl

    fps = 10
    ani = animation.FuncAnimation(fig, update, frames=Nframes,
                                  interval=1000/fps, blit=False)
    ani.save(path, writer=PillowWriter(fps=fps))
    plt.close(fig)
    print(f"Saved 2D cylindrical GIF: {path}")

# -----------------------------
# 14) End-to-end runner
# -----------------------------
def run_case(case: CaseParams, cfg: TrainConfig, make_gifs: bool = False
             ) -> Tuple[Dict[str, np.ndarray], PINNModel, Dict[str, list]]:
    fvm = run_fvm(case)
    teacher = FVMTeacher(fvm, device=device) if case.use_teacher else None

    model, hist = train(case, cfg, teacher)

    pinn_ts = eval_wall_time_series(case, model, Nt=600)

    print(f"[{case.name} | PINN FINAL] Tw_in={pinn_ts['Tw_in'][-1]:.2f} K, Tw_out={pinn_ts['Tw_out'][-1]:.2f} K")
    print(f"[{case.name} | PINN FINAL] q_in ={pinn_ts['q_in'][-1]:.2f} W/m^2, q_out={pinn_ts['q_out'][-1]:.2f} W/m^2")

    plot_wall_compare(case, fvm, pinn_ts)
    plot_flux(case, fvm, pinn_ts)

    energy_balance_check(case, fvm, pinn_ts, model=model, verbose=True)

    # (added) GIF export
    if make_gifs:
        try:
            make_gif_1d_from_pinn(case, model, Nx=140, Nframes=200,
                                  path=f"{case.name}_1d_temp.gif")
            make_gif_2d_cyl_from_pinn(case, model, Nr=140, Ntheta=180, Nframes=120,
                                      path=f"{case.name}_cyl_temp.gif")
        except Exception as e:
            print(f"[GIF skipped] {e}")

    return fvm, model, hist

def compare_two_cases(caseA: CaseParams, modelA: PINNModel,
                      caseB: CaseParams, modelB: PINNModel):
    tsA = eval_wall_time_series(caseA, modelA, Nt=400)
    tsB = eval_wall_time_series(caseB, modelB, Nt=400)

    plt.figure(figsize=(7.6, 4.0))
    plt.plot(tsA["t"], tsA["Tw_in"],  label="A PINN inner")
    plt.plot(tsA["t"], tsA["Tw_out"], label="A PINN outer")
    plt.plot(tsB["t"], tsB["Tw_in"],  label="B PINN inner")
    plt.plot(tsB["t"], tsB["Tw_out"], label="B PINN outer")
    plt.xlabel("time [s]")
    plt.ylabel("Temperature [K]")
    plt.title("PINN wall temperatures (A vs B)")
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

# -----------------------------
# 15) MAIN
# -----------------------------
if __name__ == "__main__":
    fix_seed(2025)

    presets = build_presets()
    caseA = presets["A"]
    caseB = presets["B"]

    cfg = TrainConfig(
        adam_epochs=3000,
        lr_adam=1e-3,
        use_lbfgs=True,
        lbfgs_max_iter=400,
        print_every=500,
        use_rar=True,
        rar_period=200,
        rar_candidates=20000,
        rar_add=2000,
        W_DATA0=20.0,
        teacher_off_epoch=2500,
    )

    # make_gifs=True 로 켜면 각 케이스별 GIF 2개씩 저장합니다.
    fvmA, pinnA, histA = run_case(caseA, cfg, make_gifs=True)
    fvmB, pinnB, histB = run_case(caseB, cfg, make_gifs=True)

    compare_two_cases(caseA, pinnA, caseB, pinnB)
```
